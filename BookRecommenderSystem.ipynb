### popularity based RS

```python
ratings_books = ratings.merge(books, on='ISBN')
```

```python
ratings_count = ratings_books.groupby('Book-Title').count()['Book-Rating'].reset_index()
ratings_count.rename(columns={'Book-Rating':'BookRatingCount'}, inplace=True)
ratings_count.head()
```

```python
ratings_avg = ratings_books.groupby('Book-Title').mean()['Book-Rating'].reset_index()
ratings_avg.rename(columns={'Book-Rating':'BookRating_avg'}, inplace=True)
ratings_avg.head()
```

```python
popularity_df = ratings_count.merge(ratings_avg,on='Book-Title')
popularity_df = popularity_df[popularity_df['BookRatingCount'] >= 250].sort_values('BookRating_avg',ascending=False).head(50)
popularity_df
```

```python
popular_df = popularity_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'BookRating_avg', 'Book-Author','BookRatingCount','Image-URL-M']]
popular_df.head(2)
```

## Collaborative Filtering

```python
user_rating_200 = ratings_books.groupby('User-ID').count()['Book-Rating'].reset_index()
user_rating_200.rename(columns={'Book-Rating':'BookRatingCount'}, inplace=True)
user_rating_200 = user_rating_200[user_rating_200['BookRatingCount'] >= 200]
user_rating_200.head()
eligible_user = user_rating_200['User-ID'].to_list()
print(len(eligible_user))
```

```python
filtered_user = ratings_books[ratings_books['User-ID'].isin(eligible_user)]
filtered_user.shape
```

```python
filtered_book = filtered_user.groupby('Book-Title').count()['Book-Rating'].reset_index()
filtered_book.rename(columns={'Book-Rating':'BookRatingCount'}, inplace=True)
filtered_book = filtered_book[filtered_book['BookRatingCount'] >= 50]
filtered_book.head()
eligible_book = filtered_book['Book-Title'].to_list()
```

```python
filtered_user_book = ratings_books[ratings_books['Book-Title'].isin(eligible_book)]
filtered_user_book.head()
```

```python
pt = filtered_user_book.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pt.fillna(0, inplace=True)
pt.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
sim_score = cosine_similarity(pt)
sim_score.shape
```
